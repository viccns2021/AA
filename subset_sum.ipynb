
### 1° implementação: Método guloso usando superincreasing sequences (autoral)

O(n)

```python
def findSubSeqAux(conjunto, index, numero):
    if index < 0:
        return
    else:
        if (numero >= conjunto[index]):
            # print(conjunto[index], end = " ")
            numero -= conjunto[index]
        findSubSeq(conjunto, index - 1, numero)

def findSubSeq(arr, size, sum_):
    findSubSeqAux(arr, size-1, sum_)
```

```python
A = [1, 4, 6, 13, 25, 51, 102, 204, 409, 817, 1633]
print("Resultado: " + str(findSubSeq(A, len(A), 3259)))
```

### 2° Implementação: Recursiva de tempo exponencial (autoral)

O(2^n * n)

```python
# Solução recursiva de tempo exponencial

def subsetsum_exponential(A: int, s: int, R: list = None, interator: int = 0, achei: bool = False) -> bool:
  if not R:
    R = []
  
  # O conjunto já foi "consumido" por completo? Já achamos o resultado?
  if (A and not achei):
    # item atual
    v = A[0]

    # a soma do subconjunto satisfaz s?
    if(interator + v == s):
      R.append(v)
      # print("subconjunto: "+ str(R))
      achei = True
    elif(interator + v < s):
      R_1 = R.copy()
      R_2 = R.copy()

      # adicionando o item atual a um dos subconjuntos
      R_1.append(v)

      # soma considerando o item atual
      achei = subsetsum_exponential(A[1:], s, R_1, interator + v, achei=achei)
      
      if not achei:
        achei = subsetsum_exponential(A[1:], s, R_2, interator, achei=achei)

  return achei
```

```python
A = [8, 9, 15, 15, 33, 36, 39, 45, 46, 60, 68, 73, 80, 92, 96]
print("Resultado: " + str(subsetsum_exponential(A, 183)))
```

### 3° implementação: Programação dinâmica

O(sum*n)

https://www.geeksforgeeks.org/subset-sum-problem-dp-25/

```python
# A Dynamic Programming solution for subset 
# sum problem Returns true if there is a subset of 
# set[] with sun equal to given sum 
  
# Returns true if there is a subset of set[] 
# with sum equal to given sum
def isSubsetSum(set, n, sum):
      
    # The value of subset[i][j] will be 
    # true if there is a
    # subset of set[0..j-1] with sum equal to i
    subset =([[False for i in range(sum + 1)] 
            for i in range(n + 1)])
      
    # If sum is 0, then answer is true 
    for i in range(n + 1):
        subset[i][0] = True
          
    # If sum is not 0 and set is empty, 
    # then answer is false 
    for i in range(1, sum + 1):
         subset[0][i]= False
              
    # Fill the subset table in bottom up manner
    for i in range(1, n + 1):
        for j in range(1, sum + 1):
            if j < set[i-1]:
                subset[i][j] = subset[i-1][j]
            if j >= set[i-1]:
                subset[i][j] = (subset[i-1][j] or 
                                subset[i - 1][j-set[i-1]])
      
    # uncomment this code to print table 
    # for i in range(n + 1):
    #   for j in range(sum + 1):
    #     print (subset[i][j], end =" ")
    #   print()
    # return subset[n][sum]
          
# Driver code
if __name__=='__main__':
    set = [3, 34, 4, 12, 5, 2]
    sum_ = 9
    n = len(set)
    if (isSubsetSum(set, n, sum_) == True):
        print("Found a subset with given sum")
    else:
        print("No subset with given sum")
          
# This code is contributed by 
# sahil shelangia. 
```

### Tests

Funções auxiliares

```python
def get_seed(seed=0):
  if not hasattr(get_seed, "seed"):
    get_seed.seed = seed
  get_seed.seed = get_seed.seed + 1
  return get_seed.seed

def gen_sup_inc(size: int, seed=1):
    random.seed(get_seed())
    if size < 1:
      return []
    l = [random.randint(1, 2)]
    for i in range(size):
        l.append(sum(l)+random.randint(1,2))
    return l


def test_algorithms(f: str, arr: list, sum_: int):
  start = time.time()

  size = len(arr)
  if f == "findSubSeq":
    findSubSeq(arr, size, sum_)
  if f == "subsetsum_exponential":
    subsetsum_exponential(arr, sum_)
  if f == "isSubsetSum":
    isSubsetSum(arr, size, sum_)

  end = time.time()
  return end-start 

def soma(l:list) -> int:
  r = 0
  for i in l:
    r = r + i
  return r


def plot_timeArrays(*arrs, start_x = 0):
  plt.title('Tempo de execução')
  plt.ylabel('Tempo')
  plt.xlabel('Tamanho do conjunto')
  plt.ticklabel_format(style ='scientific', axis = 'y',scilimits = (-3,3))
  plt.plot(range(start_x, len(arrs[0]) + start_x), arrs[0])
  plt.show()
  plt.title('Tempo de execução')
  plt.ylabel('Tempo')
  plt.xlabel('Tamanho do conjunto')
  plt.ticklabel_format(style ='scientific', axis = 'y',scilimits = (-3,3))
  plt.plot(range(start_x, len(arrs[1]) + start_x), arrs[1])
  plt.show()
  plt.title('Tempo de execução')
  plt.ylabel('Tempo')
  plt.xlabel('Tamanho do conjunto')
  plt.ticklabel_format(style ='scientific', axis = 'y',scilimits = (-3,3))
  plt.plot(range(start_x, len(arrs[2]) + start_x), arrs[2])
  plt.show()

  plt.title('Tempo de execução')
  plt.ylabel('Tempo')
  plt.xlabel('Tamanho do conjunto')
  plt.ticklabel_format(style ='scientific', axis = 'y',scilimits = (-3,3))

  plt.plot(range(start_x, len(arrs[0]) + start_x), arrs[0])
  plt.plot(range(start_x, len(arrs[1]) + start_x), arrs[1])
  plt.plot(range(start_x, len(arrs[2]) + start_x), arrs[2])
```

Comparação com conjuntos aleatórios

```python
timeArray1 = []
timeArray2 = []
timeArray3 = []

def run_tests():
  for i in range(20,100):
    random.seed(1)
    arr = random.sample(range(0,100), i)
    sum_ = soma(random.sample(arr, 20))

    timeArray1.append(test_algorithms("findSubSeq", arr, sum_))
    timeArray2.append(test_algorithms("subsetsum_exponential", arr, sum_))
    timeArray3.append(test_algorithms("isSubsetSum", arr, sum_))
    
  plot_timeArrays(timeArray1, timeArray2, timeArray3, start_x=20)

run_tests()
```
